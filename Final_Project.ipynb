```python
#Solving outliers
def wisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw 
```

```python
for i in df.columns:
    lw,uw=wisker(df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
    df[i]=np.where(df[i]>uw,uw,df[i])
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, root_mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
#Feature selection
X = df.drop("MEDV", axis=1)  # independent variables
y = df["MEDV"]               # target variable
```

```python
#Scaling features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Train/Test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
#Models Comparison
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import pandas as pd

models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(random_state=42)
}

results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = root_mean_squared_error(y_test, y_pred)
    
    results.append({
        "Model": name,
        "RÂ² Score": round(r2, 3),
        "MAE": round(mae, 2),
        "RMSE": round(rmse, 2)
    })

#Results in a table
results_df = pd.DataFrame(results)
print(results_df)
```

```python
#Feature Selection - Embedded Method (Random Forest Feature Importance)
best_model = RandomForestRegressor(random_state=42)
best_model.fit(X_train, y_train)
importances = pd.Series(best_model.feature_importances_, index=X.columns)
importances.sort_values(ascending=False).plot(kind='bar', title="Feature Importances from Random Forest")
plt.show()
```

```python
import joblib
joblib.dump(best_model, "random_forest_model.pkl")
```
